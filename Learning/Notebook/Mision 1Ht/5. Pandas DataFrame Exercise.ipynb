To consolidate the knowledge acquired in the use of DataFrames, I develop a simple exercise of the creation of one facture.

```python
# IMPORT LIBRARIES
import pandas as pd
import numpy as np
from random import randint
```

```python
# CREATE FICTITIOUS INFORMATION
products = np.array(['Chocorramo','Bombom','Chetos','Jugo Hit - Plastico','Gala','Quipitos','Detodito - Pequeño','Detodito - Grande','Chochitos','Menta'])
price = np.array([2200,800,1000,3000,1800,1000,2000,8000,1500,100])/2
amount = np.array([randint(1,100) for _ in range(10)])
```

```python
# CREATE DATAFRAME
df = pd.DataFrame({'product':products,'amount':amount,'price':price})
display(df)
```

```python
# CREATE A NEW COLUMN FOR THE SALE PRICE
profit = 1
sale_price = pd.Series(df['price'] * (1+profit))

# CREATE A NEW COLUMN FOR THE TOTAL PURCHASE VALUE OF EACH PRODUCT
total_cost = pd.Series(df['price'] * df['amount'])

# CREATE A NEW COLUMN TO KNOW THE PROFITS PER UNIT
profit = pd.Series(sale_price - df['price'])

# CREATE A NEW COLUMN TO KNOW THE TOTAL EARNINGS
total_profit = pd.Series(profit * df['amount'])
```

```python
# INSERT NEW DATA COLUMS
df['sale_price'] = sale_price
df['purchase_price'] = total_cost
df['unit_profit'] = profit
df['total_profit'] = total_profit

display(df)
```

```python
# CREATE A LAST ROW THAT CONTAIN THE TOTAL
df.loc[10] = df.sum(numeric_only=True)
df.loc[10, 'product'] = 'Total'

display(df)
```
