## <FONT COLOR="orange">**Exercise 1. Model Packaging**</FONT>
---
---

To evidence the new knowledge, I realize the model packaging in this section with a classification example.

```python
# IMPORT DATASET
from sklearn.datasets import load_iris

# IMPORT MODEL
from sklearn.ensemble import RandomForestClassifier

# IMPORT PACKAGING
import joblib
```

```python
# LOAD DATA AND SELECT FEATURES AND TARGET
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# CREATE MODEL
model = RandomForestClassifier()

# TRAIN MODEL
model.fit(X, y)
```

```python
# SAVE TRAINING MODEL
joblib.dump(model, 'trained_model.pkl')
```

```python
# LOAD MODEL FROM DISK
loaded_model = joblib.load('trained_model.pkl')
```

```python
# REALIZE PREDICTIONS
predictions = loaded_model.predict(X[:5])
print(f'Predictions: {predictions}')
```
