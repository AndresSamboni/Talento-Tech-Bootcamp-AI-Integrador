```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

data2x2 = [[162, 4], [11, 22]]
df2x2 = pd.DataFrame(data2x2)

print('Matriz 2x2')
display(df2x2)
```

```python
tp = df2x2.iloc[0, 0]
fp = df2x2.iloc[0, 1]
fn = df2x2.iloc[1, 0]
tn = df2x2.iloc[1, 1]

# CALCULAR METRICAS
accuracy_2x2 = (tp + tn) / (tp + fp + fn + tn)
precision_2x2 = tp / (tp + fp)
recall_2x2 = tp / (tp + fn)
f1_score_2x2 = 2 * (precision_2x2 * recall_2x2) / (precision_2x2 + recall_2x2)

# PRESENTAR LOS RESULTADOS
print(f"Accuracy: {accuracy_2x2:.4f}")
print(f"Precision: {precision_2x2:.4f}")
print(f"Recall: {recall_2x2:.4f}")
print(f"F1 Score: {f1_score_2x2:.4f}")
```
