To calculate prime values, ​​it is crucial to keep in mind that:

1. The number must be greater than 1.
2. It must have only two distinct positive divisors.
3. The divisors must be the number 1 and the number being tested.

To do this, the following code block calculates all divisors of any number to determine whether or not it is a prime number.

```python
def obtain_prime_number(number:int) -> bool:
  for i in range(2, number):
    if number % i == 0:
      return False
  return True
```

```python
# ASK THE USER FOR THE LIST LENGTH
key = True
while key:
  try:
    length = int(input('Enter the total number of prime numbers you want to obtain; note that the results will always start at 2:'))
  except:
    print(f'{Fore.LIGHTRED_EX}Enter an natural number, note that the value entered refers to the count of prime numbers delivered.{Style.RESET_ALL}')
  else:
    if length >= 1:
      key = False
    else:
      print(f'{Fore.LIGHTRED_EX}Enter a positive integer greater that 0.{Style.RESET_ALL}')

# FILL IN THE LIST WITH THE PRIME NUMBERS
prime_number = []
number = 2
while len(prime_number) < length:
  if obtain_prime_number(number):
    prime_number.append(number)
  number += 1

# DISPLAY THE ENTIRE LIST
print('The prime numbers are:')
count = 1
for i in prime_number:
  if count % 25 == 0:
    print(i)
  else:
    print(i,end=' ')
  count += 1
```

The purpose of the exercise is to allow the user to perform certain calculations on the finances of a previously established expense inventory.

To do so, the first function calculates the balance on the inventory.

```python
# BALANCE FUNCTION
def balance(account:dict) -> int:
  # GET THE DICTIONARY KEYS
  account_key = list(account.keys())
  # GET THE MONEY CAP
  base = account_key[0]
  # DROP THE TOP MONEY KEY
  account_key.pop(0)

  # CALCULATE MONTHLY EXPENSES
  expense = 0
  for i in account_key:
    if type(account[i]) == dict:
      for key, value in account[i].items():
        expense += value
    else:
      expense += account[i]

  # CALCULATE THE BALANCE
  return account[base]-expense
```

The next function, enable the modification on the inventory to allow users management more specific each part of the inventory.

```python
# MODIFY INVENTORY FUNCTION
def modify(account:dict, key=True) -> bool:
  # GET DICTIONARY KEYS
  account_key = list(account.keys())

  while key:
    # PRESENT THE FIELDS THAT THE USER CAN MODIFY
    print(f'The inventory has the following fields')
    for i in account_key:
      print(i, end='\t\t')
    field = input('\nEnter the field you want to modify: ').lower().capitalize()

    # VERIFY THE FIELD ENTERED
    if (field in account_key) and (type(account[field]) == dict):
      # IDENTIFY THE FIELD TO MODIFY
      key = modify(account[field])
    elif field in account_key:
      # GET THE NEW VALUE
      flag = True
      while flag:
        try:
          value = int(input(f'Enter the new value for {field}: '))
        except:
          print(f'{Fore.LIGHTRED_EX}Enter a natural number.{Style.RESET_ALL}')
        else:
          if value < 0:
            print(f'{Fore.LIGHTRED_EX}Enter a positive number.{Style.RESET_ALL}')
          else:
            account[field] = value
            flag = False
      return False
    else:
      print(f'{Fore.LIGHTRED_EX}The field entered does not exist.{Style.RESET_ALL}')
```

The following code block is the main function that is responsible for call the previous functions according to the user behavior.

```python
def main(inventory:dict, flag:bool=True):
  while flag:
    # DISPLAY MENU
    print(
    """
    --------------------Option menu--------------------
    1. View the inventory.
    2. Modify the inventory.
    3. Calculate the balance.
    4. Exit.
    ---------------------------------------------------
    """
    )
    # VALIDATE OPTION
    key = True
    while key:
      try:
        option = int(input('Plese, enter the option to continue:'))
      except:
        print(f'{Fore.LIGHTRED_EX}Enter a natural number between 1 and 4.{Style.RESET_ALL}')
      else:
        key = False
    # IDENTIFY OPTION
    match option:
      case 1:
        # DISPLAY INVENTORY
        print('The inventory is:')
        count_1 = 1
        for key, value in inventory.items():
          if type(value) == dict:
            print(f'{count_1}. {key}.')
            count_2 = 1
            for key2, value2 in value.items():
              print(f'\t{count_1}.{count_2}. {key2} = {value2}')
              count_2 += 1
          else:
            print(f'{count_1}. {key} = {value}')
          count_1 += 1

      case 2:
        # CAPTURE THE VALUE BEFORE MODIFICATION
        before_balance = inventory
        # APPLY THE MODIFICATION
        output = modify(inventory)
        # CALCULATE THE BALANCE
        favor = balance(inventory)
        # VALIDATE A POSITIVE BALANCE
        if favor < 0:
          inventory = before_balance
          print(f'{Fore.LIGHTRED_EX}The balance is negative, the inventory has not been modified.{Style.RESET_ALL}')
        else:
          print(f'{Fore.LIGHTGREEN_EX}The inventory has been modified.{Style.RESET_ALL}')

      case 3:
        # DISPLAY THE BALANCE OF THE INVENTORY
        print(f'The balance in favor is {balance(inventory)}')

      case 4:
        print('Exit')
        flag = False
```

```python
# INVENTORY DEFINITION
inventory = {
    'Salary': 3000000,
    'Expense': {
        'Rent':700000,
        'Food':200000,
        'Transport':720000
    },
    'Saving': {
        'Motorcicle': 200000,
        'House': 500000,
    },
    'Emergency': 250000
}

# MAIN FUNCTION EXECUTION
main(inventory)
```

 # <FONT COLOR='red'>**SET EXERCISE**</FONT>
 ---

 In this case, the objective is to practice with the sets and their operations to understand better how this structure functions.

```python
# ASK THE USER FOR TWO UNORDER LISTS
flag = True
while flag:
  print('Enter two lists, separating each item with a "," sign, but not including spaces. Note that the list does not need to be sorted:')
  try:
    set_1 = set([float(i) for i in input('List 1:').split(',')])
    set_2 = set([float(i) for i in input('List 2:').split(',')])
  except:
    print(f'{Fore.LIGHTRED_EX}Enter a list of numbers split by "," do not enter spaces or additional characters.{Style.RESET_ALL}')
  else:
    flag = False

# DISPLAY RESULTS
print('Lists were transformed into sets to work with their unique values:')
print(f'Set 1: {set_1}\t\tSet 2: {set_2}')

# MENU INTERACTION
flag = True
while flag:
  # MENU OPTIONS
  print(
  """
  --------------------Option menu--------------------
  1. Obtain the union of the two sets.
  2. Obtain the intersection of the two sets.
  3. Obtain the difference of the two sets.
  4. Obtain the symmetric difference of the two sets.
  5. Exit.
  ---------------------------------------------------
  """
  )
  # VALIDATE OPTION
  key = True
  while key:
    try:
      option = int(input('Plese, enter the option to continue:'))
    except:
      print(f'{Fore.LIGHTRED_EX}Enter a natural number between 1 and 5.{Style.RESET_ALL}')
    else:
      key = False
  # SELECT OPTION
  match option:
    case 1:
      # SETS UNION
      print(f'The union of the two sets is: {set_1.union(set_2)}')

    case 2:
      # SETS INTERSECTION
      print(f'The intersection of the two sets is: {set_1.intersection(set_2)}')

    case 3:
      # SETS DIFFERENCE
      print(f'The difference of the two sets is: {set_1.difference(set_2)}')

    case 4:
      # SYMETRIC DIFFERENCE OF SETS
      print(f'The symmetric difference of the two sets is: {set_1.symmetric_difference(set_2)}')

    case 5:
      print('Exit')
      flag = False
```
