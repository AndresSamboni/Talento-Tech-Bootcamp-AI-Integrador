```python
# IMPORT LIBRARIES
from colorama import Fore, Style
from IPython.display import clear_output
```

```python
# BIINARY SEARCH DEFINITION
def bin_search(min:int,max:int)->int:
  return (max+min)//2
```

```python
# EXECUTE THE MAIN CONTENT
if __name__ == '__main__':
  print("----------------------------------------GUESSING GAME---------------------------------------------")
  print("Hey, welcome to the game, I will guess the number you are thinking of by asking you questions.")
  print("Let's start")
  key = True
  while key:
    try:
      print("Please enter the range where the number is. Example 1-10")
      print(f"{Fore.YELLOW}Note that the range and the number must be natural numbers.{Style.RESET_ALL}")
      values = input("Range: ").split("-")
      min = int(values[0])
      max = int(values[1])
    except Exception as e:
      print(f"{Fore.RED}The values must be splitted by \"-\" and must be natural numbers{Style.RESET_ALL}")
    else:
      if(min<0 or max<0) and (max<min):
        print(f"{Fore.RED}The range entered is in the wrong format.{Style.RESET_ALL}")
      else:
        key = False
  clear_output()
  key = True
  attemps = 0
  while key:
    attemps += 1
    guess = bin_search(min,max)
    flag = True
    print("--------------------GUESSING GAME--------------------")
    while flag:
      print(f"I'm thinking about the number {guess}")
      print(f"Answer with: Correct (c), Greather (G) or Less (L)")
      answer = input().lower()
      if answer == 'c' or answer == 'correct':
        print(f"Great! Guess your number in {attemps} attempts.")
        key = False
        break
      elif answer == 'g' or answer == 'greather':
        min = guess+1
        flag = False
      elif answer == 'l' or answer == 'less':
        max = guess-1
        flag = False
      else:
        print(f"{Fore.RED}The answer entered is not valid.{Style.RESET_ALL}")
      clear_output()
```
