The notebook's objective is present the Pandas Dataframe creation using as example the student's inventory of a superior educative center.

```python
# IMPORT LIBRARIES
import pandas as pd
import numpy as np
from random import random
from random import randint
from random import seed
```

As a starting point, I create fictitious information through separate arrays.

```python
# SET SEED TO RANDOM NUMBERS
seed(42)

# SEMESTER ARRAY CREATION
semester = np.sort(np.array([randint(1,10) for _ in range(20)]))

# SUBJECT NAMES ARRAY CREATION
subjects = np.array([
  'General mathematics', 'Differential Calculus', 'Computer Science 1', 'Mechanics',
  'Electromagnetism', 'Mechanics Laboratory', 'Computer Science 2', 'Integral Calculus',
  'Vector Calculus',
  'Database management', 'Modern physics', 'Discrete mathematics', 'Inferential mathematics',
  'Transmision media',
  'Business Fundamentals', 'Telecomunications Systems Laboratory 1', 'Telecomunications Systems Laboratory 2', 'Telecomunications Systems Laboratory 3',
  'Ethics', 'Degree Thesis'
])

# ARRAY CREATION OF STUDENTS AMOUNT BY SUBJECT
count_student = np.array([randint(10,50) for _ in range(20)])

# ARRAY CREATION OF DROPOUT RATE BY SUBJECT
dropout_rate = np.array([random()*100 for _ in range(20)])

# ARRAY CREATION OF COMPLETION RATE BY SUBJECT
completion_rate = np.array([100-i for i in dropout_rate])
```

Now, it is the moment to create my Pandas Dataframe compile in one point all the previous information.

```python
# DATAFRAME CREATION
df = pd.DataFrame({
    'semester': semester,
    'subject': subjects,
    'students': count_student,
    'completition_rate': completion_rate,
    'dropout_rate': dropout_rate
})

display(df)
```
